```python
async def generate_summary_async(content, id):
    text_analytics_client_async = TextAnalyticsClientAsync(endpoint=COG_SERV_ENDPOINT, credential=AzureKeyCredential(COG_SERV_KEY))
    async with text_analytics_client_async:
        poller_async = await text_analytics_client_async.begin_analyze_actions(
            [content], 
            actions=[ExtractSummaryAction()],
            polling_interval=1)
        document_results_async = await poller_async.result()
        print(f'Generated async summary {id}')
        print()
        result_text_async = ''
        async for summary_results_async in document_results_async:
            for result_async in summary_results_async:
                if result_async.kind == "ExtractiveSummarization":
                    result_text_async += " ".join([sentence.text for sentence in result_async.sentences])  + " "
    return result_text_async
```
